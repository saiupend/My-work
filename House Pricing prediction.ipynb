```python
import pandas as pd
```

```python
df = pd.read_csv('House Prices.csv')
```

```python
df
```

```python
df = df.dropna()
```

```python
q1 = df['price'].quantile(0.25)
```

```python
q3 = df['price'].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
df = df[(df['price'] >= (q1 - 1.5 * iqr)) & (df['price'] <= (q3 + 1.5 * iqr))]
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
```

```python
df
```

```python
df[['street_name', 'street_type']] = df['street'].str.split(' ', n=1, expand=True)
```

```python
df
```

```python
df.to_csv('cleaned_House Prices.csv', index=False)
```
